````python
from google.colab import drive
from ultralytics import YOLO
import shutil
from google.colab import files

# 1. Google Drive 연결
drive.mount('/content/drive')

# 2. YOLOv8 모델 로딩 (이미 존재하는 yolo11n.pt 파일 사용)
model = YOLO('/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/yolo11n.pt')

# 3. 모델 훈련 설정
model.train(
    data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml',  # 데이터셋 경로 (data.yaml)
    epochs=100,  # 에포크 수 (성능을 높이기 위해 100으로 설정)
    batch=16,  # 배치 사이즈 (메모리 용량에 맞게 설정, 16 추천)
    imgsz=640,  # 이미지 크기 (큰 이미지 크기 선택: 640 또는 512)
    lr0=0.01,   # 초기 학습률 (작게 시작하여 점차적으로 학습)
    lrf=0.01,   # 학습률 감소 비율
    weight_decay=0.0005,  # 가중치 감쇠 (L2 정규화)
    warmup_epochs=3.0,  # 웜업 에포크 (안정적인 학습 시작)
    patience=10,  # 성능 개선 없으면 학습 중지 (early stopping)
    save_period=10,  # 주기적으로 모델 저장
    save=True,  # 훈련 중 모델 저장
    project='runs/detect/train',  # 훈련 결과 저장 폴더
    name='train_results',  # 결과 폴더 이름
    exist_ok=True,  # 이미 존재하는 폴더 덮어쓰기
    amp=True,  # 자동 혼합 정밀도(AMP) 사용
)

# 4. 훈련이 끝난 후 성능 평가
metrics = model.val(
    data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml',  # validation 데이터셋 경로
    batch=16,  # 배치 사이즈
    imgsz=640  # 이미지 크기
)

# 평가 결과 출력
print(f"Validation Results: {metrics}")

# 5. 훈련된 모델 저장 (드라이브에 모델 저장)
model_path = '/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/trained_model.pt'
model.save(model_path)
print(f"Model saved to {model_path}")

# 6. 훈련 결과 압축하여 다운로드 링크 생성
shutil.make_archive('/content/runs', 'zip', 'runs')  # 'runs' 폴더를 압축
file_path = '/content/runs.zip'  # 압축된 파일 경로

# 다운로드 링크 생성
files.download(file_path)

print("압축된 파일이 로컬로 다운로드되었습니다.")
````

네, `test`와 `train`뿐만 아니라 `valid` 폴더도 존재하는 경우, `valid` 폴더는 **검증(validation)** 데이터로 사용됩니다. 이렇게 데이터셋을 나누는 것은 **훈련**(training)과 **검증**(validation)을 분리하여, 모델이 과적합되는 것을 방지하고, 성능을 더욱 객관적으로 평가하기 위해 중요합니다.

여기서는 `train`, `valid`, `test` 폴더가 있을 때의 `data.yaml` 설정 방법과 YOLOv8 모델 학습 코드에 대해 설명하겠습니다.

### 1. `data.yaml` 파일 업데이트

`train`, `valid`, `test` 폴더가 모두 존재한다고 하셨으므로, `data.yaml` 파일을 아래와 같이 수정해야 합니다. `train` 폴더와 `valid` 폴더는 훈련과 검증에 사용되고, `test` 폴더는 최종 테스트에 사용됩니다.

```yaml
train: /content/drive/MyDrive/Cigarette Detection.v1i.yolov8/train  # 훈련 데이터 경로
val: /content/drive/MyDrive/Cigarette Detection.v1i.yolov8/valid    # 검증 데이터 경로
test: /content/drive/MyDrive/Cigarette Detection.v1i.yolov8/test    # 테스트 데이터 경로

# 클래스 이름
names:
  0: "cigarette"
```

- `train`: 훈련 데이터 경로
- `val`: 검증 데이터 경로
- `test`: 최종 테스트 데이터 경로
- `names`: 객체 탐지 클래스 이름 (여기서는 `cigarette` 하나만 정의)

### 2. YOLOv8 모델 학습 코드

`data.yaml` 파일에 따라 YOLOv8 모델을 학습하는 코드는 다음과 같습니다.

```python
from ultralytics import YOLO

# YOLO 모델 로드 (yolov8n.yaml은 작은 모델, 필요에 따라 다른 모델 사용 가능)
model = YOLO("yolov8n.yaml")

# 모델 학습
model.train(
    data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml',  # data.yaml 경로
    epochs=30,  # 에포크 수
    batch=16,   # 배치 크기
    imgsz=640,  # 이미지 크기
    early_stopping=True,  # 성능 향상이 없으면 학습 중지
)

# 학습 후 모델 평가 (검증 데이터셋에서)
results = model.val(data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml')

# 테스트 데이터셋에서 성능 평가
test_results = model.test(data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml')
print(f"Test Results: {test_results}")
```

### 3. 모델 평가 및 테스트

학습이 끝난 후, `val()` 메서드를 사용하여 검증 데이터셋에서 모델 성능을 평가할 수 있습니다. 또한 `test()` 메서드를 사용하여 테스트 데이터셋에서 최종 성능을 평가할 수 있습니다.

```python
# 검증 데이터셋에 대한 성능 평가
results = model.val(data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml')

# 테스트 데이터셋에 대한 성능 평가
test_results = model.test(data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml')
print(f"Test Results: {test_results}")
```

- **조기 종료**: 성능 향상이 멈추면 학습을 자동으로 중단하도록 설정하여 과적합을 방지할 수 있습니다.
