````python
from google.colab import drive
from ultralytics import YOLO
import shutil
from google.colab import files

# 1. Google Drive ì—°ê²°
drive.mount('/content/drive')

# 2. YOLOv8 ëª¨ë¸ ë¡œë”© (ì´ë¯¸ ì¡´ì¬í•˜ëŠ” yolo11n.pt íŒŒì¼ ì‚¬ìš©)
model = YOLO('/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/yolo11n.pt')

# 3. ëª¨ë¸ í›ˆë ¨ ì„¤ì •
model.train(
    data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml',  # ë°ì´í„°ì…‹ ê²½ë¡œ (data.yaml)
    epochs=100,  # ì—í¬í¬ ìˆ˜ (ì„±ëŠ¥ì„ ë†’ì´ê¸° ìœ„í•´ 100ìœ¼ë¡œ ì„¤ì •)
    batch=16,  # ë°°ì¹˜ ì‚¬ì´ì¦ˆ (ë©”ëª¨ë¦¬ ìš©ëŸ‰ì— ë§ê²Œ ì„¤ì •, 16 ì¶”ì²œ)
    imgsz=640,  # ì´ë¯¸ì§€ í¬ê¸° (í° ì´ë¯¸ì§€ í¬ê¸° ì„ íƒ: 640 ë˜ëŠ” 512)
    lr0=0.01,   # ì´ˆê¸° í•™ìŠµë¥  (ì‘ê²Œ ì‹œì‘í•˜ì—¬ ì ì°¨ì ìœ¼ë¡œ í•™ìŠµ)
    lrf=0.01,   # í•™ìŠµë¥  ê°ì†Œ ë¹„ìœ¨
    weight_decay=0.0005,  # ê°€ì¤‘ì¹˜ ê°ì‡  (L2 ì •ê·œí™”)
    warmup_epochs=3.0,  # ì›œì—… ì—í¬í¬ (ì•ˆì •ì ì¸ í•™ìŠµ ì‹œì‘)
    patience=10,  # ì„±ëŠ¥ ê°œì„  ì—†ìœ¼ë©´ í•™ìŠµ ì¤‘ì§€ (early stopping)
    save_period=10,  # ì£¼ê¸°ì ìœ¼ë¡œ ëª¨ë¸ ì €ì¥
    save=True,  # í›ˆë ¨ ì¤‘ ëª¨ë¸ ì €ì¥
    project='runs/detect/train',  # í›ˆë ¨ ê²°ê³¼ ì €ì¥ í´ë”
    name='train_results',  # ê²°ê³¼ í´ë” ì´ë¦„
    exist_ok=True,  # ì´ë¯¸ ì¡´ì¬í•˜ëŠ” í´ë” ë®ì–´ì“°ê¸°
    amp=True,  # ìë™ í˜¼í•© ì •ë°€ë„(AMP) ì‚¬ìš©
)

# 4. í›ˆë ¨ì´ ëë‚œ í›„ ì„±ëŠ¥ í‰ê°€
metrics = model.val(
    data='/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/data.yaml',  # validation ë°ì´í„°ì…‹ ê²½ë¡œ
    batch=16,  # ë°°ì¹˜ ì‚¬ì´ì¦ˆ
    imgsz=640  # ì´ë¯¸ì§€ í¬ê¸°
)

# í‰ê°€ ê²°ê³¼ ì¶œë ¥
print(f"Validation Results: {metrics}")

# 5. í›ˆë ¨ëœ ëª¨ë¸ ì €ì¥ (ë“œë¼ì´ë¸Œì— ëª¨ë¸ ì €ì¥)
model_path = '/content/drive/MyDrive/Cigarette Detection.v1i.yolov8/trained_model.pt'
model.save(model_path)
print(f"Model saved to {model_path}")

# 6. í›ˆë ¨ ê²°ê³¼ ì••ì¶•í•˜ì—¬ ë‹¤ìš´ë¡œë“œ ë§í¬ ìƒì„±
shutil.make_archive('/content/runs', 'zip', 'runs')  # 'runs' í´ë”ë¥¼ ì••ì¶•
file_path = '/content/runs.zip'  # ì••ì¶•ëœ íŒŒì¼ ê²½ë¡œ

# ë‹¤ìš´ë¡œë“œ ë§í¬ ìƒì„±
files.download(file_path)

print("ì••ì¶•ëœ íŒŒì¼ì´ ë¡œì»¬ë¡œ ë‹¤ìš´ë¡œë“œë˜ì—ˆìŠµë‹ˆë‹¤.")
````
